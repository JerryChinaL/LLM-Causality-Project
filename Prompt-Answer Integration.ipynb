```python
import pandas as pd
import numpy as np
import json
```

```python
response = {}
for model in ["text-ada-001","text-babbage-001","text-davinci-001","text-davinci-002","text-davinci-003"]:
    with open(model + "_response.json", "r") as file:
        response[model] = json.load(file)

with open("double_prompts2.json", "r") as file:
     double_prompts2= json.load(file)
```

```python
import pandas as pd

# Step 1: Initialize an empty DataFrame
df = pd.DataFrame(columns=["pairid", "res", "prompt", "value", "model"])

for model, model_response in response.items():
    for pairid in range(1, 109):
        for res in [1, 2]:
            item = model_response[pairid * 2 - 3 + res]
            key = next(iter(item.keys()))
            key = key[5:8] + key[-5:]
            value = next(iter(item.values()))
            prompt = double_prompts2[key]

            # Step 2: Append to the DataFrame
            df.loc[len(df)] = [pairid, res, prompt, value, model]

# Now, df contains all the rows you've appended
df.to_csv("double_prompts2.csv")
df
```

```python
response = {}
for model in ["gpt-3.5-turbo", "gpt-4"]:
    with open(model + "_response.json", "r") as file:
        response[model] = json.load(file)

with open("prompts.json", "r") as file:
     prompts= json.load(file)
```

```python
df_single_sys = pd.DataFrame(columns=["pairid", "res", "prompt", "value", "model"])

for model, model_response in response.items():
    for pairid in range(1, 109):
        item = model_response[pairid - 1]
        key = next(iter(item.keys()))
        key = str(int(key[5:8]))
        value = next(iter(item.values()))
        prompt = prompts[key]

        # Step 2: Append to the DataFrame
        df_single_sys.loc[len(df_single_sys)] = [pairid, res, prompt, value, model]

df_single_sys.to_csv("df_single_sys.csv")
df_single_sys
```
